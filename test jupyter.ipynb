Ahora solicitamos la función a analizar y su variable independiente:

```python
F=input('Ingrese la función a analizar: ')
x=input('Ingrese la variable de interés: ')
Fs=sp.sympify(F)
xs=sp.symbols(x)
Fx=fc.funcion(Fs,xs)
d=1
```

Seleccionamos el cálculo a realizar y la información necesaria para llevarlo a cabo.
